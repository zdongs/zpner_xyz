```python
import pandas as pd
```

```python
data = pd.read_csv('d:/AI/bossallex.csv',header=None)
data
```

```python
data = data.drop_duplicates()
data = data.reset_index(drop=True)
data
```

```python
import numpy as np
data = data.replace(to_replace='None', value=np.nan).dropna()
data.isnull().any()
```

```python
data = data.reset_index(drop=True)
data
```

```python
data2 = pd.read_csv('d:/AI/boss.csv',header=None)
# 合并两列
result = pd.concat([data, data2])
result
```

```python
result = result.drop_duplicates()
result = result.reset_index(drop=True)
result
```

```python
csv_path ='d:/AI/bossex.csv'
result.to_csv(csv_path,index=False,header=False)
```
