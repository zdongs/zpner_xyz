```python
import pandas as pd
df = pd.read_csv("d:/AI/bossbase/boin.csv")
```

```python
df
```

```python

def merge_row(row):
    keywords = [str(row['关键词1'])]
    
    if not pd.isna(row['关键词2']):
        keywords.append(str(row['关键词2']))
    
    if not pd.isna(row['关键词3']):
        keywords.append(str(row['关键词3']))
    
    return '\n'.join(keywords)
# 使用 apply 函数合并 2、3、4 列，并替换到 2 的位置
df['关键词1'] = df.apply(merge_row, axis=1)

# 删除原本的 3、4 列
df = df.drop(['关键词2', '关键词3'], axis=1)

df.columns = ["岗位名称", "区位", "薪资", "工作年限", "学历", "关键词", "公司名", "企业类别", "企业规模", "岗位要求","更新时间"]
df
```

```python
df.to_csv('d:/AI/bossbase/boin.csv', index=False)
```
